```python
import requests
from bs4 import BeautifulSoup

# URL de la página
url = "https://www.elearningexperience.es/zoom-meetings/nuevas-estrategias-para-detectar-e-interpretar-los-textos-generados-por-ia-en-la-educacion-media-y-superior/"

# Hacer la solicitud GET omitiendo la verificación SSL
response = requests.get(url, verify=False)

# Crear el objeto BeautifulSoup
soup = BeautifulSoup(response.content, 'html.parser')

# Buscar el div con id 'splashScreen' y extraer la URL del video
splash_screen_div = soup.find('div', {'id': 'splashScreen'})
if splash_screen_div:
    style_attr = splash_screen_div.get('style', '')
    video_url_match = re.search(r'url\("([^"]+)"\)', style_attr)
    if video_url_match:
        video_url = video_url_match.group(1)
        print("URL del video:", video_url)
    else:
        print("No se encontró la URL del video.")
else:
    print("No se encontró el div con id 'splashScreen'.")
```

```python
import requests
from bs4 import BeautifulSoup
from requests.packages.urllib3.exceptions import InsecureRequestWarning

# Desactivar las advertencias de inseguridad de requests
requests.packages.urllib3.disable_warnings(InsecureRequestWarning)

linkedins = []
profile_images = []

# URL de la página que quieres scrape
url = 'https://www.elearningexperience.es/ponentes/'

# Realizar la solicitud GET sin verificación SSL
response = requests.get(url, verify=False)
response.raise_for_status()  # Asegurarse de que la solicitud fue exitosa

# Analizar el contenido HTML con BeautifulSoup utilizando el parser 'html.parser'
soup = BeautifulSoup(response.content, 'html.parser')

# Buscar todos los enlaces con la clase 'speaker-name'
speaker_links = soup.find_all('a', class_='speaker-name')

# Extraer y mostrar los nombres y URLs de todos los ponentes
for link in speaker_links:
    speaker_name = link.text.strip()
    speaker_url = link['href']
    print(f'Nombre del ponente: {speaker_name}')
    print(f'URL del ponente: {speaker_url}')
    
    # Realizar la solicitud GET a la URL del perfil del ponente
    speaker_response = requests.get(speaker_url, verify=False)
    speaker_response.raise_for_status()
    
    # Analizar el contenido HTML del perfil del ponente
    speaker_soup = BeautifulSoup(speaker_response.content, 'html.parser')
    
    # Buscar el div con la clase 'lvsy-social'
    social_div = speaker_soup.find('div', class_='lvsy-social')
    if social_div:
        linkedin_link = social_div.find('a', href=True)
        if linkedin_link:
            linkedin_url = linkedin_link['href']
            print(f'LinkedIn del ponente: {linkedin_url}')
            linkedins.append(linkedin_url)
        else:
            print('LinkedIn no encontrado')
    else:
        print('Div lvsy-social no encontrado')
    
    # Buscar la imagen de perfil dentro del div con la clase específica
    photo_div = speaker_soup.find('div', class_='pv-top-card__non-self-photo-wrapper ml0')
    if photo_div:
        img_tag = photo_div.find('img', class_='pv-top-card-profile-picture__image--show')
        if img_tag and 'src' in img_tag.attrs:
            img_url = img_tag['src']
            print(f'Imagen de perfil: {img_url}')
            profile_images.append(img_url)
        else:
            print('Imagen de perfil no encontrada')
    else:
        print('Div de foto no encontrado')
    
    print('---')

# Imprimir listas finales de enlaces de LinkedIn e imágenes de perfil
print("Enlaces de LinkedIn de todos los ponentes:")
for linkedin in linkedins:
    print(linkedin)

print("Imágenes de perfil de todos los ponentes:")
for img in profile_images:
    print(img)
```

```python
import requests
from bs4 import BeautifulSoup
from requests.packages.urllib3.exceptions import InsecureRequestWarning

# Desactivar las advertencias de inseguridad de requests
requests.packages.urllib3.disable_warnings(InsecureRequestWarning)
desigs = []
# URL de la página principal que quieres scrape
main_url = 'https://www.elearningexperience.es/ponentes/'

# Realizar la solicitud GET sin verificación SSL
response = requests.get(main_url, verify=False)
response.raise_for_status()  # Asegurarse de que la solicitud fue exitosa

# Analizar el contenido HTML con BeautifulSoup utilizando el parser 'html.parser'
soup = BeautifulSoup(response.content, 'html.parser')

# Buscar todos los enlaces con la clase 'speaker-name'
speaker_links = soup.find_all('a', class_='speaker-name')

# Función para obtener la designación del ponente desde su página
def get_speaker_designation(speaker_url):
    response = requests.get(speaker_url, verify=False)
    response.raise_for_status()  # Asegurarse de que la solicitud fue exitosa
    soup = BeautifulSoup(response.content, 'html.parser')
    speaker_designations = soup.find_all('div', class_='speaker-designation')
    for designation in speaker_designations:
        designation_text = designation.text.strip()
        return designation_text
    return None

# Extraer y mostrar los nombres, URLs y designaciones de todos los ponentes
for link in speaker_links:
    speaker_name = link.text.strip()
    speaker_url = link['href']
    print(f'Nombre del ponente: {speaker_name}')
    print(f'URL del ponente: {speaker_url}')
    designation = get_speaker_designation(speaker_url)
    if designation:
        print(f'Designación del ponente: {designation}')
        desigs.append(designation)
    else:
        print('Designación del ponente no encontrada.')
    print('---')
```
