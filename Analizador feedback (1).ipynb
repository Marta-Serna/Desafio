```python
import pandas as pd
from nltk.corpus import stopwords
import openai

# Cargar datos
def load_data(file_path):
    return pd.read_csv(file_path)

# Preprocesar comentarios
def preprocess_comments(comments):
    stop_words = set(stopwords.words('english'))
    comments = comments.apply(lambda x: [word for word in x.split() if word not in stop_words])
    return comments

# Análisis de sentimiento usando OpenAI API
def analyze_sentiment(comment):
    response = openai.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=[
            {"role": "system", "content": "You are a psychologist."},
            {"role": "user", "content": f"Summarize the following review in a few adjectives highlighting the comments clearly: {' '.join(comment)} Summary:"}
        ],
        max_tokens=200
    )
    topics = response.choices[0].message.content.strip()
    return topics

# Función principal
def main(file_path):
    data = load_data(file_path)
    comments = data['review']  # Asegúrate de que la columna se llame 'review'
    
    # Preprocesar comentarios
    processed_comments = preprocess_comments(comments)
    
    # Análisis de sentimiento
    data['sentiment'] = processed_comments.apply(analyze_sentiment)
    
    # Mostrar estadísticas básicas
    print(data['sentiment'])
    
    # Guardar resultados
    data.to_csv('processed_feedback.csv', index=False)

# Ejecutar la función principal con el archivo de comentarios
if __name__ == "__main__":
    main(r"E:\mega\MEGAsync Downloads\reviews.csv")
```
